```python
data['START_DATE*']=data['START_DATE*'].astype('datetime64[ns]')
data['END_DATE*']=data['END_DATE*'].astype('datetime64[ns]')
```

```python
data['Time'] = data['END_DATE*']-data['START_DATE*'] #find time required by substracting the two values
data.head()
```

```python
data['Time'] = data['Time'].dt.total_seconds()/60.0
data['Time'] = data['Time'].astype('float')
data.head()
```

```python
data['KM']= data['MILES*']*1.609
```

```python
data.head()
```

```python
data.drop(['MILES*'],axis=1)
```

```python
data['speed']=(data['KM']/data['Time'])*60
```

```python
import datetime
import calendar
```

```python
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'], format="%m/%d/%Y %H:%M") 
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'], format="%m/%d/%Y %H:%M")
```

```python
hour=[]  #empty list
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in data['START_DATE*']:
    hour.append(x.hour) #adding/appending the values to above empty list
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
data['HOUR']=hour #creatig columns
data['DAY']=day
data['DAY_OF_WEEK']=dayofweek
data['MONTH']=month
data['WEEKDAY']=weekday
```
